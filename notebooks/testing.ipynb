```python
import torch
import os
from src.networks.generator import Generator
from src.eval.testing import Tester
from src.util.signals import load_chunks_pair_list
from src.util.consts import TEST_TASK_1
from src.data.filesampler import sample_filepaths
from src.util.device import set_device
```

```python
os.environ['CUDA_VISIBLE_DEVICES'] = "1"
device = set_device()
```

```python
test_paths = sample_filepaths(TEST_TASK_1, sample_rate=1)
test_chunks = load_chunks_pair_list(sampled_paths=test_paths)
print("Loaded validation dataset")
```

```python
PATH = '../output/2024-11-08_11-43-55'

generator = Generator(device)
generator.load_state_dict(torch.load(f"{PATH}/{generator.pt}"))
```

```python
tester = Tester('test')

test_chunks = load_chunks_pair_list(sampled_paths=test_paths)
tester.test(
    generator=generator,
    paths=test_paths,
    chunks=test_chunks,
    episode="testing",
    device=device,
    write_all=True,
    verbose=1
)
```
