```python
import os

from torch.utils.data import DataLoader

from src.data.dataset import SpeechData
from src.util.consts import TASK_1, TASK_2, TASK_3
from src.util.device import set_device
from src.util.lag import compute_lags

%load_ext autoreload
%autoreload 2
```

```python
#os.environ["CUDA_VISIBLE_DEVICES"] = "1"
device = set_device()
```

```python
tasks = TASK_1
dataset = SpeechData(tasks=tasks, return_chunks=False, align=False)

dataloader = DataLoader(
    dataset,
    batch_size=len(dataset),
    shuffle=False,
    collate_fn=lambda batch: zip(*batch),
)
clean_files, recorded_files, recorded_paths = next(iter(dataloader))

clean_files = clean_files
recorded_files = recorded_files

compute_lags(clean_files, recorded_files, recorded_paths, device)
```
